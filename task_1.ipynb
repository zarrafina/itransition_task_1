# Apply transformations to produce a summary table with the following fields:
- publication year,
- book_count published that year,
- average_price of books published that year in USD rounded to cents using convertion rate €1 = $1.2.

```python
import re
import json
import pandas as pd
```

```python
{:id=>10292064894005717421, :title=>"Look Homeward, Angel", :author=>"Prof. Teressa Kautzer"}
```

```python
{"id":10292064894005717421, "title":"Look Homeward, Angel", "author":"Prof. Teressa Kautzer"}
```

```python
input_file = "task_1_itransition.txt"
```

```python
with open(input_file, "r", encoding="utf-8") as f_in:
    for line in f_in:
        line = re.sub(r':(\w+)', r'"\1"', line)
        line = line.replace("=>", ":")
```

```python
data = json.loads(line)
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.dtypes
```

```python
df["price_usd"] = df["price"].where(df["price"].str.startswith("$"), "0")
df["price_euro"] = df["price"].where(df["price"].str.startswith("€"), "0")
```

```python
df
```

```python
df ["price_usd"] = df["price_usd"].str.replace("$", "", regex=False).astype(float)
df ["price_euro"] = df["price_euro"].str.replace("€", "", regex=False).astype(float)
```
